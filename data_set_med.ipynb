```python
df_ingredient = df_copy[df_copy['전문의약품'].str.contains("일반", na=False)]
df_ingredient
```

```python
#df_necessary에는 '품목기준코드','제품명','제품영문명','주성분','주성분영문','첨가제','전문의약품'에 대한 정보만 들어있음. 그리고 전문의약품 정보 중 '일반의약품'에 대한 정보만 추가되어 있음.
df_necessary = df_ingredient[['품목기준코드', '제품명', '주성분', '첨가제']]
df_necessary
```

```python
df_necessary.to_excel('medical.xlsx', index = False)
```

```python
df_list = df_necessary.values.tolist()
```

```python
df_list[0]
```

```python
results = []
Q = []

for item in df_list:
    count = 0
    # 각 항목을 초기화
    a = ""
    b = ""
    c = ""
    d = ""
    aa = ""
    
    for item2 in item:
        if count == 0:
            a = "[" + str(item2) + "] 는 품목기준코드를 의미하며"
            
            count += 1
        elif count == 1:
            b = "[" + item2 + "]는 제품명이며"
            aa = str(item2) + "에 대해 설명해주세요."
            count += 1

        elif count == 2:
            c = "해당 약품의 주성분은 [" + str(item2) + "] 이고"
            count += 1
        elif count == 3:
            d = "첨가제는 [" + str(item2) + "] 입니다."
            count += 1
        
    # 모든 문자열을 하나로 결합
    e = b + " " + c + " " + d
    results.append(e)
    Q.append(aa)

# 결과 출력
print(results[7036])
print(len(results))

print(Q[7036])
print(len(Q))
```
